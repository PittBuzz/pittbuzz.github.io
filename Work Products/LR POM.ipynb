```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

df = pd.read_excel('POM PAID-PRIO_ARNG_workingpython.xlsx',sheet_name='Sum Cost')
#df = pd.read_excel('POM PAID-PRIO_ARNG_workingpython.xlsx',sheet_name='Sum Driver')
df.columns = df.columns.str.strip()  # Removes leading/trailing spaces
```

```python
# Define years and future forecast years
years = ["2019", "2020", "2021", "2022", "2023", "2024"]
forecast_years = [str(year) for year in range(2025, 2031)]  # Forecasting till 2030

# Create a new dataframe to store forecasts
df_forecast = df.copy()

# Convert year labels to numerical values for regression
X_train = np.array([int(year) for year in years]).reshape(-1, 1)  # Reshape for sklearn
```

```python
# Run Linear Regression for each row (each ID)
for index, row in df.iterrows():
    try:
        # Extract past data and ensure it's numeric
        y_train = row[years].astype(float).values
        
        # Train Linear Regression Model
        model = LinearRegression()
        model.fit(X_train, y_train)

        # Predict future years
        X_future = np.array([int(year) for year in forecast_years]).reshape(-1, 1)
        forecast_values = model.predict(X_future)

        # Replace negative predictions with zero
        forecast_values = np.maximum(forecast_values, 0)

        # Store forecasts in dataframe
        for i, year in enumerate(forecast_years):
            df_forecast.at[index, year] = forecast_values[i]

    except Exception as e:
        print(f"Skipping row {index} due to error: {e}")
```

```python
# Save forecasted data to a new Excel file
export_file_path = "LRCost.xlsx"
df_forecast.to_excel(export_file_path, index=False)
print(f"Forecasted data saved as {export_file_path}")
```
